Corpus is in small case, now lemmatize it and remove stopwords. Only string data will be kept

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import re
wnl = WordNetLemmatizer()
```

```python
def lemmatize(text):
  text = re.sub('[^a-zA-Z]', ' ', text)
  text = text.split()
  text = [wnl.lemmatize(word) for word in text if not word in stopwords.words('english')]
  text = ' '.join(text)
  return text
```

```python
data.Text = data.Text.apply(lemmatize)
```

```python
data.iloc[1]['Text']
```

Seperating Labels from data

```python
X = data['Text']
y = data['Category']
```

Unique Labels

```python
y.unique()
```

Vectorizing the text

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=50000)
X = cv.fit_transform(X).toarray()
```

Spliting data into train and validation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train.shape
```

Training the model with Random Forrest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```
